<hr></hr>  


It is a numerical method that allows to solve mathematical problems, by means of the simulation of random variables. Although the mathematical problem does not have any relationship with random processes, it is possible to associate an artificial probabilistic model and, in this sense, we can speak of the Montecarlo method, as a universal method of calculation; in another sense, it can be classified within the experimental methods, since it does not directly solve the equations, but the stochastic process is generated and the results are measured.[1]

### Characteristics: [2]
- have been invented in the context of the development of the atomic bomb in the 1940’s.
- Are a class of computational algorithms.
- Can be applied to vast ranges of problems.
- Are not a statistical tool.
- Rely on repeated random sampling.
- Provide generally approximate solutions.
- Are used in cases where analytical or numerical solutions don’t exist or are too difficult to implement.


### Monte-Carlo methods generally follow the following steps:[2]
1. Determine the statistical properties of possible inputs
2. Generate many sets of possible inputs which follows the above properties.
3. Perform a deterministic calculation with these sets.
4. Analyze statistically the results.

## Montecarlo Integral [3]  
Monte Carlo Integration is a simple and powerful technique for approximating complicated integrals. Assume we are trying to estimate the integral of a function $f$ over some domain $D$:  
<center> $ F = \int_D^- f(\vec{x}) d\mu (\vec{x})$ </center> 
Once again, we use vector notation to indicate that f need not be one-dimensional.  
In fact, Monte Carlo techniques are used mostly for higher-dimensional integrals, or integrals that cannot be evaulated analytically.  
Assume that we have a PDF p defined over a domain D. Then the above integral is equivalent to  
<center> $ F = \int_D^-  \frac{f(\vec{x})}{p(\vec{x})} p(\vec{x}) d\mu (\vec{x})$ </center>  
When discussing continuous probability eariler, we saw that the above interval is equal to  
<center> $ E \lbrack \frac{f(\vec{x})}{p(\vec{x})}\rbrack $ </center>  
the expected value of  
<center> $\frac{f(\vec{x})}{p(\vec{x})}$ </center>  
with respect to a random variable distributed according to $p(\vec{x})$. This equality is true for any PDF on D, as long as $p(\vec{x}) \ne 0$ whenever $f(\vec{x}) \ne 0$.  
We can also estimate the value of $ E \lbrack \frac{f(\vec{x})}{p(\vec{x})}\rbrack  $ by generating a number of random samples according to p, computing $f/p$ for each sample, and finding the average of these values. As more and more samples are taken, this average is guaranteed to converge to the expected value, which is also the value of the integral. This process of averaging the value of $ \lbrack \frac{f(\vec{x})}{p(\vec{x})}\rbrack  $ for multiple random samples to estimate the value of an integral is called Monte Carlo integration [].  


### Steps to the  Montecarlo integrate [4] :    
1) Change of variable.  
	<center>$y=\frac{x-a}{b-a}$</center>

2) Determinate $dx$.   
<center>$dy=\frac{d}{dx}(\frac{x-a}{b-a})$  </center>
<center>$dy=\frac{dx\cdot (d-a) -(x-a)\cdot0}{(b-a)^2}$</center>  
<center>$dy= \frac{dx}{b-a}$  </center>
<center>$dx = (b-a)dy$</center>

3) We replace in the original integral $\int_a^b f(x) dx $, being defined in the interval $[0,1]$ of the following form:  
<center>$ \int_0^1 h(y) dy$, where:  </center>
<center>$ h(y) = f({[b-a]\cdot y+a}\cdot(b-a)) $</center> 

4) Generate $n$ pseudo-random numbers and on the computer apply the following:  
 <center>$\frac{b-a}{n} \cdot \sum_{i=0}^n f([(b-a)\cdot r + a]\cdot (b-a)) $</center>
 
5) An estimate for the error is
 <center>$ Error = (b-a) \sqrt{\frac{f^2 - (f)^2 }{n}}$ , where:    $f^2 = \frac{1}{n} \sum_{i=1}^n f^2 (x_i) $</center>

### Example of Montecarlo integral 
<center> $ f(x)= \int_{10}^{27} (x^2\cdot \sqrt{x}) dx$</center>

#### Necessary libraries and variables:

```python
import matplotlib.pyplot as pp
import numpy as np
import math
import random

valor_real_integral = 6300
```

#### Necessary methods:

```python
def fy(ri,a,b):
        x= (b-a)*ri+a
        return math.pow(x,2.0)+math.sqrt(x)

def integral(a,b):
    suma = 0.0
    ri = 0.0
    global N
    for i in range(N):
        ri = random.random()
        suma += fy(ri,a,b)
        
        error= (b-a)*(math.sqrt((math.pow(valor_real_integral,2)-suma)/N))
        total_error.append(error)
        
        if(error<10):
            N=len(total_error)
            break;
        
    res = ((b-a)/N)*suma
    return res
```

#### Approximate result of the integral:

```python
total_error = []
N = 1000000
res= integral(10,27)
print("Res: ",res)
```

#### Error decrease graph:

```python
fig, axes = pp.subplots(1, 1, figsize=(10, 6))
x=np.arange(0,len(total_error),1)
pp.plot(x, total_error,"green")
pp.ylabel('ERROR')
pp.xlabel('TOTAL ALEATORIES NUMBERS (N)')
pp.title('')
pp.grid(True)    
pp.show()
```

### graphic of the integrated area:

```python
fig, axes = pp.subplots(1, 1, figsize=(10, 6))
x = np.arange(0, 40, 1)
fx = [math.pow(i,2.0)+math.sqrt(i) for i in x]

for Yi in range(10, 27, 1):
        Zi = math.pow(Yi,2.0)+math.sqrt(Yi)
        
        for i in range(0, int(Zi), 1):
            pp.plot(Yi, i,"yellow",marker="o")
            
pp.plot(x, fx)

pp.title("Grafica función x^2 sqrt(x)")
pp.ylabel('F(x)')
pp.xlabel('X')

pp.grid(True)    
```

# Montecarlo double integral

### Steps to the  Montecarlo integrate [4]:    
The Monte Carlo method can be used to numerically approximate the value of a double integral. For a function of two variables the steps are:  
1. Pick n randomly distributed points  $(x_1,y_1),(x_2,y_2)...(x_n,y_n)$ in the rectangle.  
2. Determine the average value of the function  
<center>$ f = \frac{1}{n} \sum_{i=1}^n f(x_i,y_i)  $ </center>
3. Compute the approximation to the integral  
<center>$ \int_a^b (\int_c^d f(x,y) dy) dx = (b-a)*(d-c)*f$ </center>  
4. An estimate for the error is
<center>$ Error = (b-a)*(d-c) \sqrt{\frac{f^2 - (f)^2 }{n}}$ , where:    $f^2 = \frac{1}{n} \sum_{i=1}^n f^2(x_i,y_i) $</center>

### Example of Montecarlo double integral 
<center> $ f(x)= \int_{0}^{18} \int_{-10}^{10} 5\cdot \sqrt{x^2} + 6\cdot sin(\sqrt{y^2})dxdy$</center>

#### Necessary libraries and variables:

```python
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import matplotlib.pyplot as pp
from matplotlib import cm
import numpy as np
import random
import math

valor_real=6185
```

#### Necessary methods:

```python
def FXY(ri,a,b,c,d):
    x=(b-a)*ri+a
    y=(d-c)*ri+c
    return 5*np.sqrt(np.power(x,2.0))+6*np.sin(np.sqrt(np.power(y,2.0)))
    
def integral_doble(a,b,c,d):
    suma = 0.0
    ri = 0.0
    
    global N
    for i in range(N):
        ri = random.random()
        suma+=FXY(ri,a,b,c,d)
        
        error= (b-a) * (d-c) * math.sqrt((math.pow(valor_real,2)-suma)/N)
        total_errores.append(error)
        
        if(error<7):
            N=len(total_errores)
            break;
            
    res = (((b-a))*((d-c))*suma)/N
    return res
```

#### Approximate result of the integral:

```python
N=10000000
total_errores=[]
res = integral_doble(0,18,-10,10)
print("Res: ",res)
```

#### Error decrease graph:

```python
fig, axes = pp.subplots(1, 1, figsize=(10, 6))
x=np.arange(0,len(total_errores),1)
pp.plot(x, total_errores,"green")
pp.ylabel('ERROR')
pp.xlabel('TOTAL ALEATORIES NUMBERS (N)')
pp.title('')
pp.grid(True)    
pp.show()
```

### 3D graphic of the integrated area:

```python
%matplotlib notebook

fig = plt.figure()
ax = fig.gca(projection='3d')
X = np.arange(-20, 20, 0.25)
Y = np.arange(-20, 20, 0.25)
X, Y = np.meshgrid(X, Y)
Z = 5*np.sqrt(np.power(X,2.0))+6*np.sin(np.sqrt(np.power(Y,2.0)))
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,linewidth=0, antialiased=False)
ax.set_zlim(0., 115)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

for Xi in range(0, 18, 3):
    for Yi in range(-10, 10, 3):
        Zi = 5*np.sqrt(np.power(Xi,2.0))+6*np.sin(np.sqrt(np.power(Yi,2.0)))
        if (Zi >= 0):
            for i in range(0, int(Zi), 1):
                ax.scatter(Xi, Yi, i, marker="o", color="yellow",alpha=0.5)
        
#barra de colores 
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```
